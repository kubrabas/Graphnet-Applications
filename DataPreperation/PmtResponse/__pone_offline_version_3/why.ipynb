```python
from icecube import dataio, icetray

DAQ_TARGET = 1

def muon_lines(tree):
    return [
        line
        for line in str(tree).splitlines()
        if " MuMinus " in line or " MuPlus " in line
    ]

i3file = dataio.I3File(file_path)
daq_count = 0
target_frame = None

while i3file.more():
    frame = i3file.pop_frame()
    if frame.Stop != icetray.I3Frame.DAQ:
        continue

    if daq_count == DAQ_TARGET:
        target_frame = frame
        break

    daq_count += 1

i3file.close()

if target_frame is None:
    raise RuntimeError(f"DAQ frame {DAQ_TARGET} not found")

before = target_frame["I3MCTree"]
after = target_frame["I3MCTree_postprop"]

print(f"DAQ frame {DAQ_TARGET}")
print("========== I3MCTree ==========")
print(before)
print()
print("========== I3MCTree_postprop ==========")
print(after)

print()
print("========== quick diff ==========")
print(f"same string printout: {str(before) == str(after)}")

print()
print("I3MCTree muon lines:")
print("".join(muon_lines(before)) or "no muon lines")

print()
print("I3MCTree_postprop muon lines:")
print("".join(muon_lines(after)) or "no muon lines")

print()
print("EventProperties:")
print(target_frame["EventProperties"] if "EventProperties" in target_frame else "MISSING")

print()
print("MMCTrackList:")
print(target_frame["MMCTrackList"] if "MMCTrackList" in target_frame else "MISSING")
```
