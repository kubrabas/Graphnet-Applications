```python
from pathlib import Path
import shutil
import gzip
import sys, os
repo_root = os.path.abspath("..")
sys.path.append(repo_root)
from Helpers.extensions import I3Compressor
```

```python
SRC_DIR = Path("/project/def-nahee/kbas/POM_Response")
DST_DIR = Path("/project/def-nahee/kbas/POM_Response_GZ")

DST_DIR.mkdir(parents=True, exist_ok=True)
```

```python
i3_files = sorted(SRC_DIR.glob("*.i3"))[3874:]

print(f"Found {len(i3_files)} .i3 files to compress.")

for f in i3_files:
    gz_target = DST_DIR / (f.name + ".gz")  
    print(f"{f}  -->  {gz_target}")

    with f.open("rb") as fin, gzip.open(gz_target, "wb") as fout:
        shutil.copyfileobj(fin, fout)

```
