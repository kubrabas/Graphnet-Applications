```python
import pandas as pd
from icecube import icetray, dataio

paths = {
    "electron": electron_path,
    "muon":     muon_path,
    "tau":      tau_path,
    "NC":       NC_path,
}

rows = []
for flavor, path in paths.items():
    f = dataio.I3File(path)
    for frame in f:
        if frame.Stop == icetray.I3Frame.Physics:
            result = my_truth_extractor(frame)
            result["flavor"] = flavor
            rows.append(result)
            break
    f.close()

df = pd.DataFrame(rows).set_index("flavor")
df


```

```python
import glob
import pandas as pd
from icecube import icetray, dataio

base_paths = {
    # "electron": "/home/kbas/scratch/String340MC/Full_Geometry/Electron_PMT_Response",
    "muon": "/home/kbas/scratch/Spring2026MC/Strings_102_40m/Muon_PMT_Response",
   # "NC":       "/home/kbas/scratch/String340MC/Full_Geometry/NC_PMT_Response",
    # "tau":      "/home/kbas/scratch/String340MC/Full_Geometry/Tau_PMT_Response",
}

rows = []
for flavor, base in base_paths.items():
    files = sorted(glob.glob(base + "/**/*.i3.gz", recursive=True))
    print(f"{flavor}: {len(files)} files found")
    for filepath in files:
        try:
            f = dataio.I3File(filepath)
            while f.more():
                frame = f.pop_daq()
                if frame is None:
                    continue
                result = my_truth_extractor(frame)
                result["flavor"] = flavor
                rows.append(result)
            f.close()
        except Exception as e:
            print(f"Error in {filepath}: {e}")
            continue

df = pd.DataFrame(rows)
print(f"Total events: {len(df)}")
df
```
