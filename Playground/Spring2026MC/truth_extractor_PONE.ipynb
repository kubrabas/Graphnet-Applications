```python
import pandas as pd
from icecube import icetray, dataio

paths = {
    "electron": electron_path,
    "muon":     muon_path,
    "tau":      tau_path,
    "NC":       NC_path,
}

rows = []
for flavor, path in paths.items():
    f = dataio.I3File(path)
    for frame in f:
        if frame.Stop == icetray.I3Frame.Physics:
            result = my_truth_extractor(frame)
            result["flavor"] = flavor
            rows.append(result)
            break
    f.close()

df = pd.DataFrame(rows).set_index("flavor")
df
```
