# Event Per Flavor

```python
electron_path = "/home/kbas/scratch/String340MC/102_String/Electron_PMT_Response/electron_electron_gen_1211.i3.gz"
muon_path = "/home/kbas/scratch/String340MC/102_String/Muon_PMT_Response/muon_muon_cls_2904.i3.gz"
tau_path = "/home/kbas/scratch/String340MC/102_String/Tau_PMT_Response/tau_tau_gen_1015.i3.gz"
NC_path = "/home/kbas/scratch/String340MC/102_String/NC_PMT_Response/nc_nc_gen_029.i3.gz"





paths = {
    "electron": electron_path,
    "muon":     muon_path,
    "tau":      tau_path,
    "NC":       NC_path,
}

rows = []
for flavor, path in paths.items():
    f = dataio.I3File(path)
    for frame in f:
        if frame.Stop == icetray.I3Frame.DAQ:
            result = my_truth_extractor(frame)
            result["flavor"] = flavor
            rows.append(result)
            break
    f.close()

df = pd.DataFrame(rows).set_index("flavor")
df


```

# Many Muon Events

```python


base_paths = {
    # "electron": "/home/kbas/scratch/String340MC/102_String/Electron_PMT_Response",
    "muon": "/home/kbas/scratch/String340MC/102_String/Muon_PMT_Response",
    # "NC":       "/home/kbas/scratch/String340MC/102_String/NC_PMT_Response",
    # "tau":      "/home/kbas/scratch/String340MC/102_String/Tau_PMT_Response",
}

rows = []
for flavor, base in base_paths.items():
    files = sorted(glob.glob(base + "/**/*.i3.gz", recursive=True))
    print(f"{flavor}: {len(files)} files found")
    for filepath in files:
        try:
            f = dataio.I3File(filepath)
            while f.more():
                frame = f.pop_daq()
                if frame is None:
                    continue
                result = my_truth_extractor(frame)
                result["flavor"] = flavor
                rows.append(result)
            f.close()
        except Exception as e:
            print(f"Error in {filepath}: {e}")
            continue

df = pd.DataFrame(rows)
print(f"Total events: {len(df)}")
df
```
