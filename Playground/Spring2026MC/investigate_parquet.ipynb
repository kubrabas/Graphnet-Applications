```python
pairs = [
    ("000", features_000, truth_000),
    ("001", features_001, truth_001),
]

for name, f, t in pairs:
    f_ids = set(f["event_no"].unique().to_list())
    t_ids = set(t["event_no"].unique().to_list())

    print(f"\n{name}")
    print("feature unique:", len(f_ids))
    print("truth unique  :", len(t_ids))
    print("same ids      :", f_ids == t_ids)

    if f_ids != t_ids:
        print("in features not truth:", sorted(f_ids - t_ids)[:20])
        print("in truth not features:", sorted(t_ids - f_ids)[:20])
```

```python
for name, truth in [
    ("000", truth_000),
    ("001", truth_001),
]:
    print(f"\n{name}")

    # Her (RunID, EventID) ikilisi kaç farklı event_no'ya gidiyor?
    key_to_eventno = (
        truth
        .group_by(["RunID", "EventID"])
        .agg(pl.col("event_no").n_unique().alias("n_event_no"))
        .filter(pl.col("n_event_no") != 1)
    )

    # Her event_no kaç farklı (RunID, EventID) ikilisine gidiyor?
    eventno_to_key = (
        truth
        .group_by("event_no")
        .agg(
            pl.struct(["RunID", "EventID"]).n_unique().alias("n_run_event_pairs")
        )
        .filter(pl.col("n_run_event_pairs") != 1)
    )

    print("bad (RunID, EventID) -> event_no rows:", key_to_eventno.height)
    display(key_to_eventno)

    print("bad event_no -> (RunID, EventID) rows:", eventno_to_key.height)
    display(eventno_to_key)

    print(
        "unique (RunID, EventID):",
        truth.select(["RunID", "EventID"]).unique().height,
        "| unique event_no:",
        truth["event_no"].n_unique(),
        "| truth rows:",
        truth.height,
    )
```

```python
for name, truth in [
    ("000", truth_000),
    ("001", truth_001),
]:
    check = (
        truth
        .group_by(["RunID", "EventID"])
        .agg(
            pl.col("event_no").unique().sort().alias("event_no_values"),
            pl.col("event_no").n_unique().alias("n_event_no"),
            pl.len().alias("n_rows"),
        )
        .sort(["RunID", "EventID"])
    )

    print(f"\n{name}")
    print("all groups have exactly one event_no:", (check["n_event_no"] == 1).all())

    display(check)
```

```python
for name, truth in [
    ("000", truth_000),
    ("001", truth_001),
]:
    print(f"\n{name}")

    subrun_ids = truth["SubrunID"].unique().sort()
    subevent_ids = truth["SubEventID"].unique().sort()

    print("SubrunID unique values:")
    display(subrun_ids)

    print("SubEventID unique values:")
    display(subevent_ids)

    print("SubrunID single value:", subrun_ids.len() == 1)
    print("SubEventID single value:", subevent_ids.len() == 1)
```
