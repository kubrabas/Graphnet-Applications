```python
path = "/home/kbas/scratch/Spring2026MC/Strings_102_40m/Muon_PMT_Response/muon_muon_37999177_gen_001.i3.gz"
data_file = dataio.I3File(path)

```

```python
for i, frame in enumerate(data_file):
    if frame.Stop != icetray.I3Frame.DAQ:
        continue
    
    key1 = 'singleOMTrigger_3PMT_1DOM'
    key2 = 'singleOMTrigger_3PMT_1DOM_nonoise'
    
    if key1 not in frame or key2 not in frame:
        print(f"Frame {i}: key eksik")
        continue
    
    val1 = frame[key1]
    val2 = frame[key2]
    
    status = "SAME" if val1 == val2 else "DIFFERENT"
    print(f"Frame {i}: {status} → {key1}={val1}, {key2}={val2}")
```
