```python
import pandas as pd
from pathlib import Path

# /project/def-nahee/kbas içindeki dosya
path = Path("/project/def-nahee/kbas/i3photons_empty_counts.csv")

df = pd.read_csv(path)
print(df.shape)
print(df.head())
```

```python
df = pd.DataFrame(df)
df
```

```python
tot_empty = df["I3Photons_empty"].sum()
tot_nonempty = df["I3Photons_nonempty"].sum()

print("Total I3Photons_empty   :", int(tot_empty))
print("Total I3Photons_nonempty:", int(tot_nonempty))
print("Total (empty+nonempty)  :", int(tot_empty + tot_nonempty))

print("Mean empty per batch   :", df["I3Photons_empty"].mean())
print("Mean nonempty per batch:", df["I3Photons_nonempty"].mean())
print("Mean empty_frac        :", df["empty_frac"].mean())
```
