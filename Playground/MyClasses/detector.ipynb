### What does it expect?

- **`input_features`**: a tensor with shape **`[N, F]`**  
  - **N** = number of nodes  
  - **F** = number of features

- **`input_feature_names`**: a list of length **`F`** (the feature names)


> **Note:** The entries in `input_feature_names` must match the keys in `feature_map()` **exactly**.  
> Otherwise, you will get a `KeyError`.

> **Order matters:** `input_feature_names` defines which feature each column in `input_features` corresponds to.  
> The detector applies the transform **column-by-column** using this list, so column `i` in `input_features` is interpreted as `input_feature_names[i]`.
