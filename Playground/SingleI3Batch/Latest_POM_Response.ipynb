```python
n_daq = 0
n_physics = 0
n_other = {}

data_file = dataio.I3File(DATA_PATH)

while data_file.more():
    fr = data_file.pop_frame()
    stop = fr.Stop

    if stop == icetray.I3Frame.DAQ:
        n_daq += 1
    elif stop == icetray.I3Frame.Physics:
        n_physics += 1
    else:
        n_other[str(stop)] = n_other.get(str(stop), 0) + 1

data_file.close()
```

```python
print("DAQ:", n_daq)
print("Physics:", n_physics)
print("Other:", n_other)
```
