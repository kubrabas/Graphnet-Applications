# PMT

```python
csv_path_pmt = "/project/def-nahee/kbas/Graphnet-Applications/Playground/Datasets/summary_pmt.csv"
csv_path_i3 = "/project/def-nahee/kbas/Graphnet-Applications/Playground/Datasets/summary_i3.csv"
```

```python
df_pmt = pd.read_csv(csv_path_pmt)

display(
    df_pmt.head(5).style.set_table_attributes(
        'style="display:block; overflow-x:auto; max-width:100%; white-space:nowrap;"'
    )
)
```

#### chek if  "frames_with_EPS", "frames_with_nonempty_EPS", "frames_with_EPS_NONOISE" are all same:

```python

cols = [
    "frames_with_EPS",
    "frames_with_nonempty_EPS",
    "frames_with_EPS_NONOISE",

]

missing = [c for c in cols if c not in df_pmt.columns]
if missing:
    raise ValueError(f"Missing columns: {missing}")

all_same_mask = df_pmt[cols].nunique(axis=1).eq(1)
print(f"Rows where all 3 are identical: {all_same_mask.sum()} / {len(df_pmt)}")
```

#### chek unique "frames_with_EPS", "frames_with_nonempty_EPS", "frames_with_EPS_NONOISE", frames_with_nonempty_EPS_NONOISE per group, and count total_pulse_count_EPS_NONOISE

```python
cols_sum = [
    "frames_with_EPS",
    "frames_with_nonempty_EPS",
    "frames_with_EPS_NONOISE",
    "frames_with_nonempty_EPS_NONOISE",
    "total_pulse_count_EPS_NONOISE",
]

for c in cols_sum:
    df_pmt[c] = pd.to_numeric(df_pmt[c], errors="coerce").fillna(0)

g = (
    df_pmt.groupby("section", as_index=False)[cols_sum]
      .sum()
)

display(g)
```

# I3

```python
df_i3 = pd.read_csv(csv_path_i3)


display(
    df_i3.head(5).style.set_table_attributes(
        'style="display:block; overflow-x:auto; max-width:100%; white-space:nowrap;"'
    )
)
```

```python
# groupby section: sums + max
cols_sum = ["frames_with_nonempty_I3Photons", "frames_with_I3Photons"]
col_max = "unique_string_id_count"

# sayısala çevir (garanti)
for c in cols_sum + [col_max]:
    df_i3[c] = pd.to_numeric(df_i3[c], errors="coerce")

g = (
    df_i3.groupby("section", as_index=False)
      .agg({
          "frames_with_I3Photons": "sum",
          "frames_with_nonempty_I3Photons": "sum",
          "unique_string_id_count": "max",
      })
)

display(
    g.style.set_table_attributes(
        'style="display:block; overflow-x:auto; max-width:100%; white-space:nowrap;"'
    )
)
```
