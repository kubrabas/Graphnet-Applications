## Info:

- **Azimuth:** The azimuth is the angle of a direction measured in the horizontal (x–y) plane, typically from the +x axis.  
  **Range:** usually **0 to 2π rad** (or **0° to 360°**); sometimes **−π to π** (or **−180° to 180°**).

- **Zenith:** The zenith angle is the angle between the direction and the +z axis (straight up).  
  **Range:** **0 to π rad** (or **0° to 180°**).
